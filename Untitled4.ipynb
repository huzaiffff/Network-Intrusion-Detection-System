```python
moviematrix=df.pivot_table(index='user_id',columns='title',values="rating")
```

```python
user_rating=moviematrix['Star Wars (1977)']
```

```python
similar=moviematrix.corrwith(user_rating)
```

```python
similar.head()
```

```python
correlated=pd.DataFrame(similar,columns=['correlation'])
```

```python
def predict_movies(movie_name):
 movie_user_rating=moviematrix[movie_name]
 similar_movie=moviematrix.corrwith(user_rating)
 correlated_movie=pd.DataFrame(similar_movie,columns=['correlation'])

 correlated_movie.dropna(inplace=True)
 correlated_movie=correlated_movie.join(rating['num of ratings'])
 prediction=correlated_movie[correlated_movie['num of ratings']>100].sort_values('correlation',ascending=0)
 return prediction




```
